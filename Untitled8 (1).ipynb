```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import shap
```

```python
# Load your recruitment dataset
df = pd.read_csv("/content/recruitment_data.csv")

print( '<' + '-'*10 + ' Descriptive Statistics ' + '-'*10 + '>')
print(df.describe())

print('\n' + '<' + '-'*10 + ' Missing Values Per Column ' + '-'*10 + '>')
print(df.isnull().sum())

print('\n' + '<' + '-'*10 + ' HiringDecision Value Counts ' + '-'*10 + '>')
print(df['HiringDecision'].value_counts())

df['DistanceFromCompany'] = df['DistanceFromCompany'].astype(int)
df['Gender'] = df['Gender'].astype(int)

print('\n' +'<' + '-'*10 + ' Dataset Info ' + '-'*10 + '>')
print(df.info())
```

```python
# Set style
sns.set(style='whitegrid')

# Correlation heatmap
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap of Features')
plt.tight_layout()
plt.show()
print('\n')

plt.figure(figsize=(7, 4))
sns.barplot(x='RecruitmentStrategy', y='HiringDecision', data=df)
plt.title('Average Hiring Decision by Recruitment Strategy')
plt.ylabel('Hiring Rate')
plt.xlabel('Recruitment Strategy')
plt.tight_layout()
plt.show()
```

```python
# One-hot encode RecruitmentStrategy if needed
df = pd.get_dummies(df, columns=['RecruitmentStrategy'], drop_first=True)

# Standardize numeric columns
num_cols = ['Age', 'ExperienceYears', 'DistanceFromCompany', 'InterviewScore',
            'SkillScore', 'PersonalityScore', 'PreviousCompanies']
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Features and Target
Features = df.drop('HiringDecision', axis=1)
Target = df['HiringDecision']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(Features, Target, test_size=0.2, random_state=42)

```

```python
# Train logistic regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Predict and evaluate
lr_pred = lr_model.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, lr_pred))
print("Classification Report:\n", classification_report(y_test, lr_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, lr_pred))

# Coefficients (for feature importance)
coef_df = pd.DataFrame({'Feature': Features.columns, 'Coefficient': lr_model.coef_[0]})
print(coef_df.sort_values(by='Coefficient', ascending=False))
```

```python
# Train decision tree
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
tree_pred = tree_model.predict(X_test)

# Evaluate
print("Decision Tree Accuracy:", accuracy_score(y_test, tree_pred))
print("Classification Report:\n", classification_report(y_test, tree_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, tree_pred))

# Demographic Parity Gap
tree_test_copy = X_test.copy()
tree_test_copy['Gender'] = df.loc[X_test.index, 'Gender']
tree_test_copy['tree_pred'] = tree_pred
gap_tree = tree_test_copy.groupby('Gender')['tree_pred'].mean()
print("Demographic Parity Gap (Tree):", abs(gap_tree[0] - gap_tree[1]))
```

```python
# Use SMOTE to correct imbalance
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X_train, y_train)

# Retrain Decision Tree
tree_model_fixed = DecisionTreeClassifier(random_state=42)
tree_model_fixed.fit(X_resampled, y_resampled)
tree_pred_fixed = tree_model_fixed.predict(X_test)

# Recalculate Bias
tree_test_copy['tree_pred_fixed'] = tree_pred_fixed
gap_tree_fixed = tree_test_copy.groupby('Gender')['tree_pred_fixed'].mean()
print("Demographic Parity Gap After SMOTE:", abs(gap_tree_fixed[0] - gap_tree_fixed[1]))
```

```python
# Train random forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

# Evaluate
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print("Classification Report:\n", classification_report(y_test, rf_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_pred))
```

```python
import shap

# Convert to DataFrame if needed
X_test_df = pd.DataFrame(X_test, columns=Features.columns)

# Create SHAP explainer for Random Forest
explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_test_df)

# If shap_values is a list (which it is for classifiers), use class 1 (e.g., "Hired")
if isinstance(shap_values, list):
    shap_values = shap_values[1]

# Plot feature importance in bar format
shap.summary_plot(shap_values, X_test_df, plot_type="bar")
```

```python
# Add Gender and prediction for fairness audit
rf_test_copy = X_test.copy()
rf_test_copy['Gender'] = df.loc[X_test.index, 'Gender']
rf_test_copy['rf_pred'] = rf_pred

# Calculate Demographic Parity Gap
gap_rf = rf_test_copy.groupby('Gender')['rf_pred'].mean()
print("Demographic Parity Gap (Random Forest):", abs(gap_rf[0] - gap_rf[1]))
```

```python
from sklearn.metrics import confusion_matrix

# Compute the confusion matrices
cm_lr = confusion_matrix(y_test, lr_pred)
cm_tree = confusion_matrix(y_test, tree_pred)
cm_rf = confusion_matrix(y_test, rf_pred)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Labels and models
models = ['Logistic Regression', 'Decision Tree', 'Random Forest']
conf_matrices = [cm_lr, cm_tree, cm_rf]
categories = ['True Negative', 'False Positive', 'False Negative', 'True Positive']

# Extract counts
counts_per_model = []
for cm in conf_matrices:
    tn, fp, fn, tp = cm.ravel()
    counts_per_model.append([tn, fp, fn, tp])

counts_per_model = np.array(counts_per_model)
x = np.arange(len(categories))
width = 0.25

# Plot
plt.figure(figsize=(10, 6))
for i in range(len(models)):
    plt.bar(x + i * width, counts_per_model[i], width, label=models[i])

plt.xlabel('Confusion Matrix Categories')
plt.ylabel('Count')
plt.title('Confusion Matrix Comparison Across Models')
plt.xticks(x + width, categories)
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
